## Import Dependencies

```python
import pandas as pd
import numpy as np
import ast
from sklearn.model_selection import train_test_split
# from sklearn.metrics import accuray, precison
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

## Load Data

```python
df = pd.read_csv('loan_approval_dataset.csv')
df.head()
```

```python
df[' loan_status'] = df[' loan_status'].apply(lambda x:x.strip())
```

```python
df[' loan_status'].value_counts()
```

```python
df.isnull().sum()
```

## Brain Storming

```python
# Columns to remove: loan_id, 
df = df.drop(columns=['loan_id'])
```

```python
df.describe().round(2)
```

```python
df[' education'].value_counts()
df[' self_employed'].value_counts()
```

```python
df[' education'].replace({' Graduate':1, " Not Graduate":0}, inplace=True)
df[' self_employed'].replace({' Yes':1, " No":0}, inplace=True)
```

```python
df.head()
```

```python
new_df = pd.DataFrame({
    'no_of_dependents': df['no_of_dependents'],
    'education': df[' education'],
    'self_employed': df[' self_employed'],
    'income_annum': df[' income_annum'],
    'loan_amount': df[' loan_amount'],
    'loan_term': df[' loan_term'],
    'cibil_score': df[' cibil_score'],
    'residential_assets_value': df[' residential_assets_value'],
    'commercial_assets_value': df[' commercial_assets_value'],
    'luxury_assets_value': df[' luxury_assets_value'],
    'bank_asset_value': df[' bank_asset_value'],
    'loan_status': df[' loan_status'],
})
```

```python
new_df.head()
```

```python
x = new_df.drop('loan_status', axis=1)
y = new_df['loan_status']
```

## Train_Test Split

```python
x_train,x_test, y_train,y_test = train_test_split(x,y, random_state=42, test_size=0.20)
display(x_train.shape)
display(x_test.shape)
display(y_train.shape)
display(y_test.shape)
```

## Model Train (Logistic Regression)

```python
def evaluate_model(modelName):
    logistic_model = modelName
    logistic_model.fit(x_train,y_train)
    y_test_pred = logistic_model.predict(x_test)
    recall = recall_score(y_test, y_test_pred, pos_label='Approved')
    accuracy = accuracy_score(y_test, y_test_pred)
    precision = precision_score(y_test, y_test_pred, pos_label='Approved')
    return {'model':modelName, 'accuracy': accuracy, 'precision':precision, 'recall':recall}
```

```python
list1 = []
LogisticRegression
models = [LogisticRegression(),DecisionTreeClassifier(), RandomForestClassifier() ]
for model in models:
    list1.append(evaluate_model(model))
```

## Training Decision Tree

```python
decision_model = DecisionTreeClassifier()
decision_model.fit(x_train,y_train)
y_test_pred = decision_model.predict(x_test)
```

```python
recall = recall_score(y_test, y_test_pred, pos_label='Approved')
accuracy = accuracy_score(y_test, y_test_pred)
precision = precision_score(y_test, y_test_pred, pos_label='Approved')
```
